```python
# For neural networks.
import keras as kr
# For data frames.
import pandas as pd
# For numerical arrays.
import numpy as np
# For preprocessing data.
import sklearn.preprocessing as pre
# For splitting data sets.
import sklearn.model_selection as mod
# For whitening.
import sklearn.decomposition as dec
# import matplotlib.pyplot for plotting  
import matplotlib.pyplot as plt 
#magic function, renders the plots in this notebook
%matplotlib inline
from sklearn.model_selection import train_test_split
```

```python
powerproduction = pd.read_csv("https://raw.githubusercontent.com/ianmcloughlin/2020A-machstat-project/master/dataset/powerproduction.csv")
powerproduction
```

```python
#Divide the data set into X and y where X (windspeed) is the independent variable and y (power generated) is the dependent variable.
#This breakdown is applicable for both regression styles
speed = powerproduction.speed.to_numpy()
y = powerproduction.power.to_numpy()
#even though you've only one input value it must be reshaped when using scikit learn
X = speed.reshape(-1, 1)

#Split the data set observations into 70 % train 30% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

#### Linear

```python
# Keras code adapted from Regression in keras lecture Ian McLoughlin
# Create a neural network with one neuron.
#sequential model
#weight initialised as one
#bias is zero
#Adam is "A Method for Stochastic Optimization"
model = kr.models.Sequential()
model.add(kr.layers.Dense(1, input_shape=(1,), activation="linear", kernel_initializer='ones', bias_initializer='zeros'))
model.compile(optimizer='adam', loss='mean_squared_error')
```

something like this:
the fit method on the model- takes the x and corresponding y values.

Compares the X_train dataset to the y_train dataset, passed the x values through the neural network, looking for
what the neural network computes the y value to be. Compare it using the mean squared error.

MSE is a risk function that assesses the quality of a prediction.

```python

# Train the neural network on our training data.
#popping the fit function onto the model produced above
#a simple code which belies the complex machinations beneath!
model.fit(X_train, y_train, epochs=500)
#epoch the number of times each X value passes through the Neural network, an attempt is made to improve the predictions
#at each cycle
#The function produces a log to the screen with information produced by the loss function
#this helps assess whether the model is a good one or not - the number should reduce.
#perhaps the dataset can be preprocessed to improve results
#Also, every additional time the fitting is processed, the result improves.
#that is a big deal!
```

```python
#Visualise it
plt.scatter(X,y,color='seagreen')

plt.plot(X_test,model.predict(X_test),color='darkgrey')

plt.title('Using Keras Neural Network - Linear')
plt.xlabel('Power')
plt.ylabel('Speed')
plt.tight_layout()
plt.subplots_adjust(top=0.88)
plt.suptitle('Power Production')






```

Evaluation is a process during development of the model to check whether the model is best fit for the given problem and corresponding data.

```python
#model.evaluate(X_test, y_test)
score = model.evaluate(X_test, y_test, verbose=0)
```

Prediction is the final step and our expected outcome of the model generation. Keras provides a method, predict to get the prediction of the trained model.

Here, all arguments are optional except the first argument, which refers the unknown input data. The shape should be maintained to get the proper prediction.

We call evaluate on the model with the testing data  verbosity off, so we dont see output on the screen.

```python
polymodel = kr.models.Sequential()#model style and compilation has not changed from before
#NOTE - 2 layers of one and 50
#50 layers this time is the hidden layer, input shape is the same, the rest are different too
#sigmoid is the curve shape that is a good fit for the powerproduction dataset
#glorot uniform is a good one to use when there is a single hidden layer
polymodel.add(kr.layers.Dense(50, input_shape=(1,), activation='sigmoid', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
polymodel.add(kr.layers.Dense(1, activation='linear', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))
#lr = learning rate, passing that is so the model doesn't train too quickly!
polymodel.compile(kr.optimizers.Adam(lr=0.001), loss='mean_squared_error')
```

```python

# Fit the data.
# batch_size : Integer or NULL. Number of samples per gradient update. If unspecified, batch_size will default to 32.
#pass ten in at a time
polymodel.fit(X_train, y_train, epochs=500, batch_size=10)
#the fitting process takes notably longer because it is a more complex network
```

```python
#Visualise it, with the test data instead
#plt.scatter(X,y,color='seagreen')
#plt.plot(X, y, 'seagreen', label="actual data")
plt.plot(X, y, 'seagreen', label="actual data")
#plt.plot(X_test,polymodel.predict(X_test),color='darkgrey')
plt.plot(X_test, polymodel.predict(X_test), 'k', label="keras prediction")
plt.title('Using Keras Neural Network - sigmoid')

plt.xlabel('Power')
plt.ylabel('Speed')
plt.tight_layout()
plt.subplots_adjust(top=0.88)
plt.suptitle('Power Production')
plt.legend();
```
